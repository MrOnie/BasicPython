#### 1.	Lea el fichero proporcionado.

```python
nombre_archivo = 'FicheroALeer.txt'
fichero = open(nombre_archivo, 'r') # Abrir fichero. Lectura.
fichero_str = fichero.read() # Lectura del fichero
print(fichero_str) # Comentar esta línea para que no se imprima el fichero
```

#### 2.	Cuente el número de líneas del fichero.

```python
fichero.seek(0) # Change stream position (offset). 0: start of stream
numero_lineas = len(fichero.readlines()) # readlines(): regresa una lista de lineas del fichero
print("El número de líneas es: ",numero_lineas)
```

#### 3.	Cuente el número de veces que aparece el artículo «el» en el fichero (nota: elemento contiene los caracteres «el» pero no debe contar puesto que no contiene a el artículo «el».

```python
palabra_a_buscar = 'el'
# Abrir archivo para lectura
with open(nombre_archivo) as archivo:
    contador_palabra = 0
    for linea in archivo:
        palabras = linea.split() # separar línea en palabras, por espacio en blanco
        for palabra in palabras:
            palabra = palabra.lower().strip(".,") # cambiar a minúscula y quitar puntuación
            if palabra == palabra_a_buscar:
                contador_palabra += 1

print("Número de veces que aparece la palabra <<"+palabra_a_buscar+">>: "+str(contador_palabra))
```

#### 4.	Elimine las tildes del fichero y convierta todas las letras a minúscula.

```python
fichero_str_minus = fichero_str.lower() # Convierte todo a minúsculas

# Función para reemplazar vocales tildadas con vocales sin tildar
def quitar_tildes(string):
    replacements = ( # vocales a reemplazar
        ('á', 'a'),
        ('é', 'e'),
        ('í', 'i'),
        ('ó', 'o'),
        ('ú', 'u'),
    )
    # Reemplazo
    for a, b in replacements:
        string = string.replace(a, b)
    return string

minuscula_sin_tildes = quitar_tildes(fichero_str_minus)
```

#### 5.	Crea un fichero de texto con el nombre «resultado.txt» que contendrá en la primera línea el número de líneas del fichero, en la segunda línea el número de veces que aparece el artículo «el», y a partir de la tercera línea el texto original todo en minúsculas y sin tilde.

```python
#import os
file = open('resultado.txt', 'w')
file.write("Número de lineas: "+str(numero_lineas))
file.write("\nNúmero de veces que aparece el artículo <<el>>: "+str(contador_palabra)+"\n")
file.write(minuscula_sin_tildes)
file.close()
```

#### Carga en el entorno el «poblacionMunicipios.csv» proporcionado por el profesor respetando las tildes y caracteres españoles.

```python
# Librerías necesarias
import numpy as np
import pandas as pd
```

```python
# Leemos el csv con pandas
nombre_csv = 'poblacionMunicipios.csv'
poblacionMunicipios = pd.read_csv(nombre_csv)
# Display funciona únicamente en los Notebooks!!!
display(poblacionMunicipios) # Comentar esta línea para que no se imprima la tabla
```

#### Elimina las filas correspondientes a municipios sin población.

```python
poblacionMunicipios = poblacionMunicipios[poblacionMunicipios.Poblacion > 0]
# Display funciona únicamente en los Notebooks!!!
display(poblacionMunicipios) # Comentar esta línea para que no se imprima la tabla
```

#### Imprime el total de la población de todos los municipios.

```python
poblacion_total = np.sum(poblacionMunicipios.Poblacion)
print("La población total de todos los municipios es: "+str(poblacion_total))
```

#### Crea un DataFrame de Pandas donde la primera columna sean todas las provincias que aparecen en el archivo, la segunda columna el número total de habitantes por provincia, la tercera la desviación típica en el número total de habitantes por provincia y la cuarta el número total de municipios por provincia.

```python
# Agrupación de provincias y cálculo del número total de habitantes por provincia
habitantes_por_provincia = poblacionMunicipios.groupby(['Provincia'], sort=False, as_index=False).sum()
# Desviación típica en el número total de habitantes por provincia
std_por_provincia = poblacionMunicipios.groupby(['Provincia'], sort=False).std()
# Número total de municipios por provincia
municipios_por_provincia = poblacionMunicipios.groupby(['Provincia'], sort=False).size()
# https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html
```

```python
resultado = pd.DataFrame({
    'Provincia': habitantes_por_provincia.Provincia,
    'Habitantes_Prov': habitantes_por_provincia.Poblacion,
    'STD_Hab_Prov': np.array(std_por_provincia.Poblacion),
    'Municipios_Prov': np.array(municipios_por_provincia)
})

display(resultado)
```

#### Empleando el código de provincia y el código de municipio, cruza los dos datasets comentados hasta el momento. El resultado final debe ser un DataFrame.

#### Genera un CSV llamado «faltan.csv» con todos los códigos postales que estén en «CP_Municipios.csv» pero que no tengan en «poblacionMunicipios.csv» ningún municipio con el que poder cruzar. Si todo cruza correctamente el archivo «faltan.csv» estará vacío. Se puede decir que «faltan.csv» son los códigos postales que se han perdido en el cruce a la hora de considerar la población.

```python
faltan = pd.merge(CP_Municipios, poblacionMunicipios, on=['CodProvincia', 'CodMunicipio'], how='outer', indicator=True) 
# True: añade la columna "-merge" con información de la fuente de cada fila 
faltan = faltan[faltan['_merge']=='left_only']
# “left_only” for observations whose merge key only appears in ‘left’ DataFrame
display(faltan)
faltan.to_csv('faltan.csv', index=False)
```

#### Agrupa la información por código postal. Se quiere un DataFrame con las siguientes columnas:
- Código postal.
- Número de municipios que tienen dicho código postal asignado. (Dividido también por provincias)
- Población: se calculará como la suma de la población de todos los municipios que incluyen a dicho código postal.
- Provincia: a la que está asignada el código postal. 

```python
# Agrupación de código postal y población por código postal
# Agrupado por código postal Y por provincia (hay casos que se repite el código postal en distintas provincias)
cruce_datasets = pd.merge(poblacionMunicipios, CP_Municipios, on=['CodProvincia', 'CodMunicipio', 'Municipio'])
CP = cruce_datasets.groupby(['CP', 'CodProvincia'], sort=False, as_index=False).sum()
# Número de municipios por código postal
num_municipios = cruce_datasets.groupby(['CP', 'CodProvincia'], sort=False, as_index=False).size()
# Provincia a la que está asignada el código postal
lista_prov = cruce_datasets.groupby(['CodProvincia', 'Provincia'], sort=False, as_index=False).sum()
provincia_CP = pd.merge(CP.CodProvincia, lista_prov[['CodProvincia', 'Provincia']], on=['CodProvincia'])
```

```python
resultado2 = pd.DataFrame({
    'CP': CP.CP,
    'Num_Municipios': np.array(num_municipios),
    'Poblacion': CP.Poblacion,
    'Provincia': provincia_CP.Provincia
})

display(resultado2)
```
