#### 1. Crea una función en Python que calcule el mínimo común múltiplo de dos números.
#### 2. Crea una función en Python que calcule el máximo común divisor de dos números.

En la bibliografía consultada, el algoritmo para calcular el mínimo común múltiplo (**mcm**) hace uso de la función para calcular el Máximo Común Divisor (**MCD**).
Por lo tanto creamos ambas funciones en el mismo bloque de código:

```python
#Función mínimo común múltiplo (mcm)
def mcm(numero1, numero2):
    return numero1 * numero2 / MCD(numero1, numero2)
    
#Función Máximo Común Divisor (MCD)
def MCD(numero1, numero2):
    while(numero1 != numero2):
        if(numero1 > numero2):
            numero1 -= numero2
        else:
            numero2 -= numero1
    return numero1
```

Creamos una función para solicitar el ingreso de un número. La función creada está validada para que permita el ingreso de números **enteros** (si es flotante, se considera solamente la parte entera) y **positivos** (mayores a 0).

```python
#Funcion para solicitar el ingreso de un entero positivo
#Como parámetro de la función se debe incluir el texto del número que se solicita (ej. número 1, número de puntos, etc)
def ingreso_int(texto_solicitud):
    entero_positivo = 0              #Inicializacón para que el algoritmo se ejecute, por lo menos, una vez
    while(entero_positivo <= 0):     #Validación para que sea número positivo mayor a cero
        entero_positivo = int(float(input("Ingresa el "+str(texto_solicitud)+" :")))     #Validación para tomar solamente la parte entera (en caso de flotantes)
    print(" El "+str(texto_solicitud)+" es:", entero_positivo)
    return entero_positivo
```

Solicitamos el ingreso de los números y procedemos a observar el resultado:

```python
#Solicitud de ingreso de los números
print("***Los números ingresados deben ser enteros positivos***")
num1 = ingreso_int("número 1")
num2 = ingreso_int("número 2")

#Usando las funciones definidas previamente
print("***Resultados:***")
print("El minimo común múltiplo (mcm) entre los números ingresados es:", int(mcm(num1,num2)))
print("El Máximo Común Divisor (MCD) entre los números ingresados es:", int(MCD(num1,num2)))
```

#### 4. Basándote en datos aleatorios, crea un Scatter Plot con Mapplotlib. Explica para qué pueden ser útiles estos gráficos.

```python
#Librería(s) necesaria(s)
import numpy as np
import matplotlib.pyplot as plt

# Semilla para reproducir el resultado
np.random.seed(20000000)

#Solicitud de número de puntos a graficar
n_puntos = ingreso_int("número de puntos")

#Función para la generación de los puntos aleatorios
def generaDatos(numero_datos):
    x = np.random.rand(numero_datos)
    y = np.random.rand(numero_datos)
    color = np.random.rand(numero_datos)
    area = (15 * np.random.rand(numero_datos))**2
    return x, y, color, area

#Puntos aleatorios
x, y, color, area = generaDatos(n_puntos)

#Definición del gráfico y sus atributos
plt.scatter(x, y, s=area, c=color, alpha=0.3)    #aplha: intensidad
plt.title('Scatter Plot de '+ str(n_puntos) +' valores aleatorios')
plt.xlabel('eje X')
plt.ylabel('eje Y')
plt.colorbar();     #barra de colores lateral
plt.show()
```

Los diagramas de puntos o "scatter plots" sirven para representar gráficamente la relación entre datos multidimensionales (en este caso, 4: x, y, color y área). Es decir, funcionan perfectamente para presentar visualmente un set de datos en ejercicios de Aprendizaje Automático de clasificación o de agrupamiento, por ejemplo.

#### 5. Explica cómo combinar 4 gráficos en una única figura (subplot).

Comenzamos por importar la librería _matplotlib.pyplot_ y las demás necesarias.

```python
#Librería(s) necesaria(s)
import numpy as np
import matplotlib.pyplot as plt
```

Definimos los parámetros básicos de nuestras gráficas. En este caso vamos a generar datos aleatorios, por lo que solamente necesitamos el número de puntos a graficar:

```python
# Semilla para reproducir el resultado
np.random.seed(19680801)
puntos = ingreso_int("número de puntos")
```

Utilizamos la función _subplot_, la cual recibe como parámetros la dimensión de la malla de subgráficos y el número del elemento:
- número de columnas
- número de filas
- número del elemento

Por lo que, si se tiene atención al momento de estructurar nuestros gráficos, no se debería tener mayor inconveniente:

```python
#Definición del gráfico y sus atributos
#plt.subplot(abc), en donde a y b son las dimensiones de la malla (axb) de subgráficos, y c es el número del elemento
plt.subplot(221)     #Malla de 2x2, elemento 1
x1, y1, color1, area1 = generaDatos(puntos)
plt.scatter(x1, y1, s=area1, c=color1, marker=">")
plt.xticks([]), plt.yticks([])
plt.title('subplot(2,2,1)')

plt.subplot(222)     #Malla de 2x2, elemento 2
x2, y2, color2, area2 = generaDatos(puntos)
plt.scatter(x2, y2, s=area2, c=color2, marker=(5, 0))
plt.xticks([]), plt.yticks([])
plt.title('subplot(2,2,2)')

plt.subplot(223)     #Malla de 2x2, elemento 3
x3, y3, color3, area3 = generaDatos(puntos)
plt.scatter(x3, y3, s=area3, c=color3, marker='+')
plt.xticks([]), plt.yticks([])
plt.title('subplot(2,2,3)')

plt.subplot(224)     #Malla de 2x2, elemento 4
x4, y4, color4, area4 = generaDatos(puntos)
plt.scatter(x4, y4, s=area4, c=color4, marker=(5, 1))
plt.xticks([]), plt.yticks([])
plt.title('subplot(2,2,4)')

plt.show()
```

Desde luego, hay maneras más "complejas" de presentar 4 gráficos combinados. Por ejemplo, mediante una malla, no tan sencilla como la presentada anteriormente. En este caso, una malla de 2 x 3:

```python
#Definición del tamaño de la malla
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)

#Al ser una matriz, se debe tomar en cuenta que los índices comienzan en 0, 
# y ya no se especifica el número del elemento, sino los rangos dentro de la matriz
plt.subplot(grid[0, 0])
plt.scatter(x1, y1, s=area1, c=color1, marker=">")
plt.subplot(grid[0, 1:])
plt.scatter(x2, y2, s=area2, c=color2, marker=(5, 0))
plt.subplot(grid[1, :2])
plt.scatter(x3, y3, s=area3, c=color3, marker='+')
plt.subplot(grid[1, 2])
plt.scatter(x4, y4, s=area4, c=color4, marker=(5, 1))

plt.show()
```
